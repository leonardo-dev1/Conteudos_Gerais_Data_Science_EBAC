```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.figure_factory as ff
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
```

Durante a aula nossa variável a ser prevista (churn) continha apenas 2 categorias, a base de vocês contém mais. O Naive Bayes pode ser aplicado para problemas de classificação com múltiplas classes da mesma forma que para problemas de classificação binária. O Naive Bayes é um algoritmo de classificação probabilístico que calcula a probabilidade de uma amostra pertencer a cada classe e seleciona a classe com a maior probabilidade como a previsão final.
Em resumo, o Naive Bayes pode ser aplicado da mesma maneira para problemas de classificação com múltiplas classes, e os mesmos princípios se aplicam em termos de treinamento, avaliação e aplicação do modelo.

# 1) Comece carregando as bases de treino (X e y) e teste (X e y).
Verifique se o número de linhas condiz, se as variáveis estão corretas sendo apenas a de score para y e as demais nas bases de X e por último, se Y está balanceada no teste.

```python
x_train = pd.read_csv('Dados/01. x_train_balanced.csv')
x_test = pd.read_csv('Dados/03. x_test.csv')
y_train = pd.read_csv('Dados/02. y_train_balanced.csv')
y_test = pd.read_csv('Dados/04. y_test.csv')
```

# 2) Aplique o algoritmo de Naive Bayes aos dados de treinamento.

```python
# Instância do Classificador
naive_credit_score = GaussianNB()

# Treinamento do Modelo
naive_credit_score.fit(x_train, y_train)
```

# 3) Faça a avaliação do modelo com os dados de treinamento.
Traga a acurácia, recall e plote a matriz de confusão. Não se esqueça de avaliar com suas palavras o desempenho do modelo, interpretando as métricas.

Dica: Para calcularmos o recall em classificação multi classe precisamos usar o atributo macro:
recall = recall_score(y_train, y_pred_train, average='macro')

```python
# Fazendo previsões nos dados de Treino
y_pred_train = naive_credit_score.predict(x_train)

# ACURÁCIA
train_accuracy = accuracy_score(y_train, y_pred_train)
print("Acurácia:", train_accuracy)

# RECALL
train_recall = recall_score(y_train, y_pred_train, average = 'macro')
print("Recall:", train_recall)
```

```python
# Matriz de Confusão
conf_matrix = confusion_matrix(y_train, y_pred_train)

# Definindo os nomes de classes
class_names = ['Low', 'Average', 'High']

# Matriz de Confusão
fig = ff.create_annotated_heatmap(
    z = conf_matrix,
    x = class_names,
    y = class_names,
    colorscale = 'Blues',
    showscale = True
)

# Estilizando
fig.update_layout(
    title = 'Matriz de Confusão',
    xaxis_title = 'Rótulos Previstos',
    yaxis_title = 'Rótulos Verdadeiros',
    font = dict(
        family = 'Arial',
        size = 12,
        color = 'black'
    ),
    template = 'plotly'
)

fig.show()
```

## **Insights**

O modelo acertou **98% de todas as previsões** (acurácia), consequentemente o recall também acompanhou esse valor, com **uma média 98% de acerto em todas as categorias** de Score de Crédito nas bases de Treinamento.

Analisando a Matriz de Confusão, podemos notar que o **modelo teve "maior dificuldade" em acertar a categória "Low"**, apontando 2 registros como "Average", e uma maior **facilidade de prever registros "High"**, apontando corretamente 83 de 84 registros.

# 4) Aplique o modelo aos dados de teste e realize a avaliação dos resultados, da mesma forma que fez acima. Não se esqueça de avaliar com as suas palavras e comparar o desempenho da base treino com a teste.

```python
# Fazendo previsões nos dados de Teste
y_pred_test = naive_credit_score.predict(x_test)

# ACURÁCIA
test_accuracy = accuracy_score(y_test, y_pred_test)
print("Acurácia:", test_accuracy)

# RECALL
test_recall = recall_score(y_test, y_pred_test, average = 'macro')
print("Recall:", test_recall)
```

```python
# Matriz de Confusão
conf_matrix = confusion_matrix(y_test, y_pred_test)

# Definindo os nomes de classes
class_names = ['Low', 'Average', 'High']

# Matriz de Confusão
fig = ff.create_annotated_heatmap(
    z = conf_matrix,
    x = class_names,
    y = class_names,
    colorscale = 'Blues',
    showscale = True
)

# Estilizando
fig.update_layout(
    title = 'Matriz de Confusão',
    xaxis_title = 'Rótulos Previstos',
    yaxis_title = 'Rótulos Verdadeiros',
    font = dict(
        family = 'Arial',
        size = 12,
        color = 'black'
    ),
    template = 'plotly'
)

fig.show()
```

## **Insights**

O Modelo também apresentou um **desempenho excelente nos dados de Teste**, **acertando 97% de todas as previsões feitas** (acurácia) e um **acerto médio de 94% por categória** de Score de Crédito (recall)

Analisando a Matriz de Confusão, podemos notar que a "dificuldade" de acertar a categória "Low" apresentada nas bases de Treino também se manteve nas bases de Teste, apontando 1 registro "Low" como "Average". Novamente houve um destaque na categória "High", com o modelo acertando 100% de todos os 29 registros "High".

Com base nos dados de Teste, a príncipio podemos afirmar que o modelo conseguiu generalizar muito bem os dados de Treino e conseguiu apresentar uma ótima assertividade nos dados de Teste
